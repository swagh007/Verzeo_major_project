```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as lib 
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
Cars= pd.read_excel('Data_Train.xlsx')

Cars= Cars.dropna()



Cars[['Company', 'Model']]= Cars['Name'].str.split(' ', 1, expand= True)
Cars[['mileage', 'Unit1']]= Cars['Mileage'].str.split(' ',1,expand=True)
Cars[['engine', 'Unit2']]= Cars['Engine'].str.split(' ',1,expand=True)
Cars[['power', 'Unit3']]= Cars['Power'].str.split(' ',1,expand=True)
Cars=Cars.drop(['Name', 'Unit1', 'Unit2', 'Unit3', 'Mileage', 'Engine', 'Power', 'Model'], axis=1)
loc=pd.get_dummies(Cars.Location)
fuel=pd.get_dummies(Cars.Fuel_Type)
transmition=pd.get_dummies(Cars.Transmission)
owner=pd.get_dummies(Cars.Owner_Type)
company=pd.get_dummies(Cars.Company)
Cars= pd.concat([Cars,loc,fuel,transmition,owner,company], axis= 'columns')
Cars= Cars.drop(['Ambassador','Lamborghini','Location', 'Pune', 'Petrol', 'Fuel_Type', 'Transmission', 'Automatic','Owner_Type', 'First', 'Company'], axis='columns')
Cars.replace('null', '0', inplace=True)


Cars_reg= linear_model.LinearRegression()
X= Cars.drop('Price', axis='columns')
X
y= Cars.Price
Cars_reg.fit(X,y)

Cars.corr().style.background_gradient(cmap='coolwarm')

#As is vissible from the data correlation table that the price of the used cars have negative correlation(blue) with every parrameter showing decline in market value of a car after use






```
